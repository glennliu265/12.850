```python

import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# FEM-UW1, Set Courant Number
u  = 1
dt = 1
dx = 1


# Note need to find a better way to specify stop
kdx = np.arange(0,math.pi+0.001,math.pi/24)

#nu  = u*dt*dx

nu_all = np.arange(0,1.6,0.5)
nu_all = [0.25,0.5,1,1.5]
```

```python

ax  = plt.subplot(111, polar=True)
#plt.hold(True)

lam = np.zeros((len(nu_all)+1,len(kdx)))
for n in range(0,len(nu_all)):
    print(str(n))
    nu = nu_all[n]

    a2 = np.power(1 - nu + nu * np.cos(kdx),2)
    b2 = np.power(nu * np.sin(kdx),2)

    lam[n,:] = np.sqrt(a2 + b2)
    

    #Plot
    c   = ax.plot(kdx,lam[n,:],label=r"$\nu$ = %.2f" % nu_all[n])
 




ax.set_thetamin(0)
ax.set_thetamax(180)
ax.set_thetagrids(range(0, 180, 15)) # set the gridlines
ax.set_rticks([0,0.5,1,1.5,2])
ax.set_title('Amplitude-Factor Modulus')
ax.set_xlabel(r'|$\lambda$|',FontSize=12,position=(0.5,1)) # Math text between $$
#ax.legend(mode="expand",ncol=1)
ax.legend(bbox_to_anchor=(-0.1, 0.05, 1, 1),ncol=2)
plt.savefig('HW5_Amplitude_Modulus.png', dpi=600) # Need to turn off plt.show for it to work
```
